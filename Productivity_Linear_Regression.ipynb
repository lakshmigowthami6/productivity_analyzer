**Encoding Categorical Data**

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['High school or below','Graduate','Undergraduate'],['Less than 2','2–4','4–6','6–8','8-10','More than 10'],['Less than 10 minutes','10–30 minutes','30–60 minutes','More than 1 hour'],['Ignore them until my task is completed','Turn off notifications altogether','Check them briefly and resume my work','Spend time interacting with the notifications'],['No, i do not use them','Yes, but i did not find them of any help','Yes, they are extremely helpful']])
```

```python
ordinal_cols = ['Education Level', 'Average Screen Time', 'Attention Span', 'Notification Handling', 'Usage of Productivity Apps']
oe.fit(X_train[ordinal_cols])
```

```python
X_train[ordinal_cols] = oe.transform(X_train[ordinal_cols])
X_test[ordinal_cols] = oe.transform(X_test[ordinal_cols])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
import numpy as np
ohe = OneHotEncoder(sparse_output=False, drop='first', dtype=np.int32, handle_unknown='ignore')
```

```python
X_train_new = ohe.fit_transform(X_train[['Age Group','Gender','Occupation','Device','Screen Activity','App Category','Screen Time Period','Environment','Work Strategy']])
X_test_new = ohe.transform(X_test[['Age Group','Gender','Occupation','Device','Screen Activity','App Category','Screen Time Period','Environment','Work Strategy']])
```

```python
nominal_cols = ['Age Group','Gender','Occupation','Device','Screen Activity','App Category','Screen Time Period','Environment','Work Strategy']
X_train_processed = X_train.drop(columns=nominal_cols)
X_test_processed = X_test.drop(columns=nominal_cols)

# Create DataFrames from the one-hot encoded arrays with appropriate column names
ohe_feature_names = ohe.get_feature_names_out(nominal_cols)
X_train_ohe_df = pd.DataFrame(X_train_new, columns=ohe_feature_names, index=X_train_processed.index)
X_test_ohe_df = pd.DataFrame(X_test_new, columns=ohe_feature_names, index=X_test_processed.index)

# Concatenate the ordinal-encoded features with the one-hot encoded features
X_train_final = pd.concat([X_train_processed, X_train_ohe_df], axis=1)
X_test_final = pd.concat([X_test_processed, X_test_ohe_df], axis=1)

print("Shape of X_train_final:", X_train_final.shape)
print("Shape of X_test_final:", X_test_final.shape)
X_train_final.head()
```

### Feature Scaling and Logistic Regression Model

Before training a new model, it's good practice to scale the features. This helps models converge faster and perform better, especially for models that use distance calculations or gradient descent. We will use `StandardScaler` to standardize the features, then train a `LogisticRegression` model, which is a common choice for classification tasks.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

scaler = StandardScaler()

# Fit the scaler on the training data and transform both training and testing data
X_train_final_scaled = scaler.fit_transform(X_train_final)
X_test_final_scaled = scaler.transform(X_test_final)

print("Features scaled successfully!")
```

```python
# Initialize and train the Logistic Regression model
logistic_model = LogisticRegression(random_state=42, max_iter=200) # Increased max_iter for convergence
logistic_model.fit(X_train_final_scaled, y_train)

print("Logistic Regression model trained.")
```

```python
# Make predictions on the scaled test set
y_pred_logistic = logistic_model.predict(X_test_final_scaled)

# Evaluate the accuracy of the Logistic Regression model
accuracy_logistic = accuracy_score(y_test, y_pred_logistic)
print(f"Accuracy of Logistic Regression model: {accuracy_logistic:.2f}")
```

### Decision Tree Classifier

Let's try training a `DecisionTreeClassifier` to see if a different model architecture can improve the accuracy. Decision trees are non-linear models that can be effective for classification tasks.

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize the Decision Tree Classifier
decision_tree_model = DecisionTreeClassifier(random_state=42)

# Train the model on the scaled training data
decision_tree_model.fit(X_train_final_scaled, y_train)

print("Decision Tree Classifier trained.")
```

```python
# Make predictions on the scaled test set
y_pred_tree = decision_tree_model.predict(X_test_final_scaled)

# Evaluate the accuracy of the Decision Tree Classifier
accuracy_tree = accuracy_score(y_test, y_pred_tree)
print(f"Accuracy of Decision Tree Classifier: {accuracy_tree:.2f}")
```

### Random Forest Classifier

Let's try a `RandomForestClassifier`. This is an ensemble learning method that builds multiple decision trees during training and outputs the class that is the mode of the classes (classification) or mean prediction (regression) of the individual trees. Random Forests generally have better predictive accuracy than a single decision tree and mitigate overfitting.

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize the Random Forest Classifier
random_forest_model = RandomForestClassifier(random_state=42)

# Train the model on the scaled training data
random_forest_model.fit(X_train_final_scaled, y_train)

print("Random Forest Classifier trained.")
```

```python
# Make predictions on the scaled test set
y_pred_rf = random_forest_model.predict(X_test_final_scaled)

# Evaluate the accuracy of the Random Forest Classifier
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Accuracy of Random Forest Classifier: {accuracy_rf:.2f}")
```
